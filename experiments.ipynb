

ANN IMPLEMENTATION


```python
#sequential network 
#hidden neuron - dense we use 
#Activation function - sigmoid, tanh, relu , leaky relu - hidden layer - relu 
# op - sigmoid o
# optimizer - back propogation - responsible the weights, 
# loss function - reduce gradient descent 
#Matrix - accuracy , mse mean absolute error 
# logs in a folder and display the board - tensor 
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

#inputs - x_train.shape 
dense to create hidden layer - input - x_train.shape[1] 

```python
X_train.shape[1] 
```

```python
##Build ANN 
model =Sequential([
    Dense(64,activation = 'relu',input_shape=(X_train.shape[1],)),  ##Hiddel Layer 1 - connected with input layer 
    Dense(32,activation='relu'),                                    ##Hiddel Layer 2
    Dense(1,activation='sigmoid') 
    ]
)
```

```python
model.summary()
```

```python
import tensorflow 
opt= tensorflow.keras.optimizers.Adam(learning_rate=0.01) 
los= tensorflow.keras.losses.binary_crossentropy
```

```python
##Compile model 
model.compile(optimizer=opt,loss=los,metrics=['accuracy']) 
```

```python
##set up tensor board - visaluze loss while calling model 


log_dir = "logs/fit/" +datetime.datetime.now().strftime("%Y%m%d-%H%M%S")  
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
##set up early stopping - train for any no of epochs, forward and backward prop, 
early_stopping_callback = EarlyStopping(monitor = 'val_loss',patience =10,restore_best_weights=True)
```

```python
###Training the model 
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100, 
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```

```python
model.save('model.h5') 
```
